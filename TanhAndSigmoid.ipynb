```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the tanh function
def tanh(x):
    return np.tanh(x)

# Generate values for x-axis
x = np.linspace(-5, 5, 1000)

# Calculate sigmoid and tanh values
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y_sigmoid, label=r'$\sigma(x) = \frac{1}{1 + e^{-x}}$', color='blue')
plt.plot(x, y_tanh, label=r'$tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$', color='red')

# Adding title and labels
plt.title('Sigmoid and Tanh Functions', fontsize=18)
plt.xlabel('x', fontsize=16)
plt.ylabel('Function Value', fontsize=16)
# Adding clear black lines for x and y axes
plt.axhline(0, color='black', linewidth=1.5)
plt.axvline(0, color='black', linewidth=1.5)
plt.xticks( fontsize=14)
plt.yticks(fontsize=14)

# Adding grid and legend
plt.grid(True)
plt.legend(fontsize=16)
plt.savefig('TANSIGM.png')

# Show the plot
plt.show()
```
