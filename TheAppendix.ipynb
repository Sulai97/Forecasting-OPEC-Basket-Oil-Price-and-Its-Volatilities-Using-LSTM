```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV data
df = pd.read_csv('test.csv')

# Convert Training Time from seconds to minutes
df['Training Time (minutes)'] = df['Training Time'] / 60

# Create a pivot table for the heatmap
pivot = df.pivot_table(values='Training Time (minutes)', index='LSTM Units', columns='Epochs', aggfunc=np.mean)

# Create the heatmap
plt.figure(figsize=(12, 8))
heatmap = sns.heatmap(pivot, annot=True, fmt='.1f', cmap='YlOrRd', cbar_kws={'label': 'Training Time (minutes)'})

# Set the font size of the color bar label manually after creating the heatmap
colorbar = heatmap.collections[0].colorbar
colorbar.ax.yaxis.label.set_size(14)

# Set titles and labels with increased font sizes
plt.title('Training Time Heatmap: LSTM Units vs Epochs', fontsize=18)
plt.xlabel('Epochs', fontsize=16)
plt.ylabel('LSTM Units', fontsize=16)

# Set xticks and yticks font size to 14
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Convert total time from seconds to hours
total_time_hours = 87624.13472 / 3600

# Add an annotation inside a box in the top left corner
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
plt.text(0.05, 0.95, f'Total Time: {total_time_hours:.2f} hours', transform=heatmap.transAxes, fontsize=13,
         verticalalignment='top', bbox=props)

# Show the plot
plt.tight_layout()
plt.savefig('The Total time Test.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'LSTM Units': [10, 50, 100, 200, 500],
    'Average Test RMSE': [1.609148458, 1.570413227, 1.55278039, 1.555859899, 1.572100765],
    'Sum of Time': [1194.190296/60, 1980.236176/60, 2539.370769/60, 3048.405396/60, 9875.644374/60]  # Converted to minutes
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting the Average Test RMSE as a line graph
ax1.plot(df['LSTM Units'], df['Average Test RMSE'], color='blue', marker='o', label='Average Test RMSE', linewidth=2)
ax1.set_xlabel('LSTM Units', fontsize=14)
ax1.set_ylabel('Average Test RMSE', fontsize=14, color='blue')
ax1.tick_params(axis='y', labelcolor='blue', labelsize=14)
ax1.tick_params(axis='x', labelsize=14)

# Create a second y-axis for the Sum of Time
ax2 = ax1.twinx()
bars = ax2.bar(df['LSTM Units'], df['Sum of Time'], color='red', label='Sum of Time', linewidth=2)

# Make the bars thicker by setting their linewidth and edge color
for bar in bars:
    bar.set_linewidth(10)
    bar.set_edgecolor('red')

ax2.set_ylabel('Sum of Time (Minutes)', fontsize=14, color='red')
ax2.tick_params(axis='y', labelcolor='red', labelsize=12)

# Add a title and a grid
plt.title(' Average Test RMSE and Total  Time for 65% Split and Different Epochs', fontsize=16)
ax1.grid(True)

# Add legends
ax1.legend(loc='upper left', fontsize=12)
ax2.legend(loc='lower right', fontsize=12)

# Show the plot
plt.tight_layout()
plt.savefig('Appendix1.png')
plt.show()
```
